Ahora, vamos a descargar los datos históricos de las acciones de Tesla para los últimos 5 años.


```python
import yfinance as yf

# Descargar los datos de las acciones de Tesla
tesla_stock = yf.Ticker("TSLA")
tesla_stock_data = tesla_stock.history(period="5y")

# Mostrar las primeras filas de los datos descargados para verificar
tesla_stock_data.head()
```

El código anterior descarga los datos históricos de Tesla y muestra las primeras filas para que podamos verificar que los datos se han descargado correctamente.


# Pregunta 2: Extracción de datos de ingresos de Tesla utilizando Web Scraping
En esta sección, utilizaremos `requests` y `BeautifulSoup` para realizar web scraping y obtener los datos de ingresos de Tesla desde Yahoo Finance.

Primero, debemos asegurarnos de que las librerías `requests` y `BeautifulSoup` estén instaladas.


# Instalar BeautifulSoup4 y requests si no están instalados
%pip install beautifulsoup4 requests


Ahora, vamos a extraer los datos de ingresos de Tesla.


```python
import requests
from bs4 import BeautifulSoup

# URL de las finanzas de Tesla en Yahoo Finance
url = "https://finance.yahoo.com/quote/TSLA/financials?p=TSLA"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Extraer la tabla de ingresos
table = soup.find_all("div", class_="D(tbr)")
tesla_income_data = {}
for row in table:
    columns = row.find_all("div", class_="D(ib)")
    if len(columns) > 1:
        label = columns[0].get_text()
        value = columns[1].get_text()
        tesla_income_data[label] = value

tesla_income_data
```

El código anterior realiza web scraping para extraer los datos de ingresos de Tesla de Yahoo Finance y los almacena en un diccionario. Aquí mostramos el contenido del diccionario para verificar los datos extraídos.

# Pregunta 4: Extracción de datos de ingresos de GameStop utilizando Web Scraping
Ahora vamos a realizar web scraping para obtener los datos de ingresos de GameStop desde Yahoo Finance.


```python
import requests
from bs4 import BeautifulSoup

# URL de las finanzas de GameStop en Yahoo Finance
url = "https://finance.yahoo.com/quote/GME/financials?p=GME"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Extraer la tabla de ingresos
table = soup.find_all("div", class_="D(tbr)")
gamestop_income_data = {}
for row in table:
    columns = row.find_all("div", class_="D(ib)")
    if len(columns) > 1:
        label = columns[0].get_text()
        value = columns[1].get_text()
        gamestop_income_data[label] = value

gamestop_income_data
```

El código anterior realiza web scraping para extraer los datos de ingresos de GameStop y los almacena en un diccionario. Aquí mostramos el contenido del diccionario para verificar los datos extraídos.


# Pregunta 5: Tablero de acciones e ingresos de Tesla
En esta sección, crearemos un cuadro de mando (dashboard) para visualizar el precio de las acciones de Tesla junto con sus ingresos.


```python
import matplotlib.pyplot as plt

# Crear gráficos para Tesla
plt.figure(figsize=(14, 7))

# Gráfico del precio de las acciones
plt.subplot(2, 1, 1)
plt.plot(tesla_stock_data.index, tesla_stock_data['Close'], label="Tesla Stock Price", color='blue')
plt.title("Tesla Stock Price Over Time")
plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.legend()

# Gráfico de ingresos
plt.subplot(2, 1, 2)
plt.bar(tesla_income_data.keys(), [float(value.replace(",", "")) for value in tesla_income_data.values()], label="Tesla Income", color='green')
plt.title("Tesla Income Over Time")
plt.xlabel("Year")
plt.ylabel("Income (USD)")
plt.legend()

plt.tight_layout()
plt.show()
```

El código anterior crea dos gráficos: uno para el precio de las acciones de Tesla y otro para sus ingresos. Se muestran juntos en un solo dashboard para facilitar la comparación.
